# Twitter Engagement Based Portfolio

## 1. Load Twitter Engagement data

- Load twitter engagement dataset, calculate engagement ratio, and filter out stocks with low activity

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import yfinance as yf
import os
plt.style.use('ggplot')

folder = 'E:/Downloads'

df = pd.read_csv(os.path.join(folder, 'sentiment_data.csv'))

df['date'] = pd.to_datetime(df['date'])

df = df.set_index(['date', 'symbol'])

df['engagement'] = df['twitterComments']/df['twitterLikes']

df = df[(df['twitterLikes']>20)&(df['twitterComments']>10)]

df
```

## 2. Calculate Average Engagement on the Monthly Level Per Stock

```python
monthlyEngagementdf = (df.reset_index('symbol').groupby([pd.Grouper(freq='M'), 'symbol']) [['engagement']].mean())

monthlyEngagementdf['rank'] = (monthlyEngagementdf.groupby(level=0)['engagement'].transform(lambda x: x.rank(ascending=False)))

monthlyEngagementdf
```

## 3. Select Top 5 Most Engaged With Stocks For Each Month

```python
topFivedf = monthlyEngagementdf[monthlyEngagementdf['rank']<=5].copy()

topFivedf = topFivedf.reset_index(level=1)

topFivedf.index = topFivedf.index+pd.DateOffset(1)

topFivedf = topFivedf.reset_index().set_index(['date', 'symbol'])

topFivedf.head(20)
```

## 4. Form a portfolio at the start of each month based on the engagement

- Creating a dictionary that matches the first day's of the month with the stocks with the most engagement

```python
days = topFivedf.index.get_level_values('date').unique().tolist()

fixedDays = {}

for d in days:
    
    fixedDays[d.strftime('%Y-%m-%d')] = topFivedf.xs(d, level=0).index.tolist()
    
fixedDays
```

## 5. Download new stock prices for selected stocks

```python
stocks_list = df.index.get_level_values('symbol').unique().tolist()

newdf = yf.download(tickers=stocks_list, start='2021-01-01', end='2023-03-01')
```

## 6. Calculate monthly portfolio returns

```python
returnsdf = np.log(newdf['Adj Close']).diff().dropna()

portfoliodf = pd.DataFrame()

for startDay in fixedDays.keys():
    
    endDay = (pd.to_datetime(startDay)+pd.offsets.MonthEnd()).strftime('%Y-%m-%d')
    
    cols = fixedDays[startDay]
    
    tempdf = returnsdf[startDay:endDay][cols].mean(axis=1).to_frame('Portfolio Return')
    
    portfoliodf = pd.concat([portfoliodf, tempdf], axis=0)
    
portfoliodf
```

# Compare NASDAQ to Twitter Engagement Based Returns

```python
qqqdf = yf.download(tickers='QQQ', start='2021-01-01', end='2023-03-01')

qqqRet = np.log(qqqdf['Adj Close']).diff().to_frame('Nasdaq Return')

portfoliodf = portfoliodf.merge(qqqRet, left_index=True, right_index=True)

portfoliodf
```

```python
import matplotlib.ticker as mtick

cumulativeReturn = np.exp(np.log1p(portfoliodf).cumsum()).sub(1)

cumulativeReturn.plot(figsize=(16,6))

plt.title('Twitter Engagement Based Portfolio Vs. Nasdaq Returns Over Time')

plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1))

plt.ylabel('Return')

plt.show()
```
